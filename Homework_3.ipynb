```python
import numpy as np
import matplotlib.pyplot as plt

inf = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt') 
inf
```

Первая задача

```python
fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')
Hares = inf[:, 1]
Lynxes = inf[:, 2]
Carrots = inf[:, 3]
Years = inf[:, 0]

plt.plot(Years, Hares, label='Hares', c="g")
plt.plot(Years, Lynxes, label='Lynxes', c='r')
plt.plot(Years, Carrots, label='Carrots', c='b')

plt.xlabel('Year', fontsize=16)
plt.ylabel('Population', fontsize=16)
plt.xticks(np.arange(1900, 1921, step=1))
plt.grid()
plt.legend()
plt.show()
```

Вторая задача

```python
fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')
Hares = inf[:,1]
Lynxes = inf[:,2]
Carrots = inf[:,3]
GoodDots = np.any(inf >= 50000, axis = 1)
GoodDotsX = inf[GoodDots,0]
GoodDotsY = np.max(inf[GoodDots, 1:], axis = 1)
plt.plot(Years, Hares, label='Hares', c="g")
plt.plot(Years, Lynxes, label='Lynxes', c='r')
plt.plot(Years, Carrots, label='Carrots', c='b')
plt.axhline(y=50000, c ='black', linestyle='--')
plt.scatter(GoodDotsX, GoodDotsY, s = 150, c = "m")
plt.xlabel('Year', fontsize=16)
plt.ylabel('Population', fontsize=16)
plt.xticks(np.arange(1900, 1922, step=1))
plt.grid()
plt.legend()
plt.show()
```

Третья задача

```python
fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')
MaxPopInd = np.argmax(inf, axis = 0)[1:]
MaxHares = inf[MaxPopInd[0], 1]
MaxHaresYear = inf[MaxPopInd[0], 0]
MaxLynxes = inf[MaxPopInd[1], 2]
MaxLynxesYear = inf[MaxPopInd[1], 0]
MaxCarrots = inf[MaxPopInd[2], 3]
MaxCarrotsYear = inf[MaxPopInd[2], 0]

plt.scatter(MaxHaresYear, MaxHares, s = 150, c = "m")
plt.scatter(MaxLynxesYear, MaxLynxes, s = 150, c = "m")
plt.scatter(MaxCarrotsYear, MaxCarrots, s = 150, c = "m")

Hares = inf[:,1]
Lynxes = inf[:,2]
Carrots = inf[:,3]
plt.plot(Years, Hares, label='Hares', c="g")
plt.plot(Years, Lynxes, label='Lynxes', c='r')
plt.plot(Years, Carrots, label='Carrots', c='b')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Population', fontsize=16)
plt.xticks(np.arange(1900, 1922, step=1))
plt.grid()
plt.legend()
plt.show()



```

Четвертая задача

```python
fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')
ax = plt.subplot()

ax.bar(Years, Hares, label = 'Hares', color = "b")
ax.bar(Years, Lynxes, bottom = Hares,
       label='Lynxes', color="g")
plt.xlabel('Year', fontsize=16)
plt.ylabel('Population', fontsize=16)
plt.xticks(np.arange(1900, 1922, step=1))
plt.grid()
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
HarePercent = Hares/(Hares+Lynxes)*100
LynxPercent = Lynxes/(Hares+Lynxes)*100

ax.bar(Years, HarePercent, label = 'Hares', color = "b")
ax.bar(Years, LynxPercent, bottom = HarePercent,
       label='Lynxes', color="g")
plt.xlabel('Year', fontsize=16)
plt.ylabel('Population', fontsize=16)
plt.title('Ratio of hares and lynxes')
plt.xticks(np.arange(1900, 1922, step=1))
plt.grid()
plt.legend()
plt.show()
```

Пятая задача

```python
fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')

print("Hares mean is:", np.mean(Hares), "; Hares std is:", np.std(Hares))
print("Lynxes mean is:", np.mean(Lynxes), "; Lynxes std is:", np.std(Lynxes))
print("Carrots mean is:", np.mean(Carrots), "; Carrots std is:", np.std(Carrots))
fig, ax = plt.subplots(1, 3, figsize=(12, 7))

ax[0].hist(Hares, color = 'g')
ax[0].set_title('Hares')
ax[0].axvline(x=np.mean(Hares), c = 'black', linestyle = "--")
ax[1].hist(Lynxes, color = 'r')
ax[1].set_title('Lynxes')
ax[1].axvline(x=np.mean(Lynxes), c = 'black', linestyle = "--")
ax[2].hist(Carrots, color = 'b')
ax[2].set_title('Carrots')
ax[2].axvline(x=np.mean(Carrots), c = 'black', linestyle = "--")

plt.show()
```

Шестая задача

```python
def name(num, a):
    if num == 1:
        return 'L'
    else:
        return 'H'
    
MaxPopulation = np.argmax(inf[:, 1:3], axis = 1)
VectFunc =  np.vectorize(name)
MaxPopulation = VectFunc(MaxPopulation, 0)

plt.pie([np.count_nonzero(MaxPopulation == "H"), np.count_nonzero(MaxPopulation == "L")], labels=['Hares', 'Lynxes'], colors=["g", "b"], shadow=True) 
plt.show()
```

Седьмая задача

```python
Hare = inf [:-1, 1]
Help =  inf [1:, 1]
Delta = - Hare + Help
Lynxe = inf [:-1, 2]
CorCoef = np.corrcoef(Delta, Lynxe)[0, 1]
print('Correlation coefficient is:', CorCoef)

fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')
X = np.linspace(1900, 1919, 20)
plt.plot(X, Delta, label='Delta')
plt.plot(X, Lynxe, label='Lynxes', c='r')

plt.xlabel('Year', fontsize=16)
plt.ylabel('Num', fontsize=16)
plt.xticks(np.arange(1900, 1920, step=1))
plt.grid()
plt.legend()
plt.show()
```

Восьмая задача

Девятая задача

```python
fig = plt.figure(figsize = (15,15))
fig.patch.set_facecolor('xkcd:white')
Hare = inf [:-2, 1]


Help =  inf [1:-1, 1]
Delta = - Hare + Help
Lynxe = inf [:-2, 2]
coef = Delta / Lynxe
Coef = np.mean(coef)
DeltaHares = Coef * inf[-2, 2]
NewInf = inf
NewInf[-1, 1] = int(DeltaHares + NewInf[-2, 1])


Lynxe = inf [:-2, 2]
Help =  inf [1:-1, 2]
Delta = - Lynxe + Help
Hare = inf [:-2, 1]
coef = Delta / Hare
Coef = np.mean(coef)
DeltaLynxes = Coef * inf[-2, 1]
NewInf[-1, 2] = int(DeltaLynxes + NewInf[-2, 2])
NewInf[-1, 3] = np.mean(NewInf[:-1, 3])

X = np.linspace(1900, 1920, 21)
plt.plot(X, Hares, label='Hares')
plt.plot(X, Lynxes, label='Lynxes', c='r')
plt.plot(X, Carrots, label='Carrots', c='orange')

plt.xlabel('Year', fontsize=16)
plt.ylabel('Population', fontsize=16)
plt.xticks(np.arange(1900, 1921, step=1))
plt.scatter(1920, NewInf[-1,1], s = 100, c = "black")
plt.scatter(1920, NewInf[-1,2], s = 100, c = "black")
plt.scatter(1920, NewInf[-1,3], s = 100, c = "black")
plt.grid()
plt.legend()
plt.show()
```
